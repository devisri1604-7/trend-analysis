**PHASE:5**

**HYPOTHESIS TESTING**

**ANOVA**

```python
from scipy.stats import f_oneway

groups = [group["Satisfaction"].dropna()
          for name, group in df.groupby("AI_Tool")]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0")
    print("There is a significant difference in satisfaction among AI tools.")
else:
    print("Fail to Reject H0")
    print("There is no significant difference in satisfaction among AI tools.")
```

**INTERPRETATION**

```python
if p_value < 0.05:
    print("\nBusiness Interpretation:")
    print("Different AI tools generate different levels of user satisfaction.")
    print("Organizations should focus on the tools with higher satisfaction scores.")
else:
    print("\nBusiness Interpretation:")
    print("User satisfaction is similar across all AI tools.")
    print("Organizations can choose AI tools based on cost, features, and business requirements.")
```

**PHASE:6**

**ORGANIZATION SEGMENTATION**

```python
import numpy as np

conditions = [
    (df["Productivity_Gain"] >= 20) & (df["AI_Investment"] >= 50000),
    (df["Adoption_Year"] <= 2020),
    (df["Adoption_Year"] >= 2023),
    (df["Satisfaction"] >= 8),
    ((df["Cost_Savings"] / df["AI_Investment"]) >= 1.5)
]

categories = [
    "AI Leaders",
    "Early Adopters",
    "Slow Adopters",
    "High Satisfaction Organizations",
    "High ROI Organizations"
]

df["Organization_Segment"] = np.select(
    conditions,
    categories,
    default="Others"
)

print(df[["Organization","Organization_Segment"]].head())
```

```python
print(df["Organization_Segment"].value_counts())
```

```python
import matplotlib.pyplot as plt

df["Organization_Segment"].value_counts().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Organization Segmentation")
plt.xlabel("Segment")
plt.ylabel("Number of Organizations")
plt.xticks(rotation=45)

plt.show()
```

```python
segment_productivity = df.groupby(
    "Organization_Segment"
)["Productivity_Gain"].mean()

print(segment_productivity)
```

```python
segment_productivity.plot(kind="bar")

plt.title("Average Productivity Gain by Segment")
plt.xlabel("Segment")
plt.ylabel("Productivity Gain")

plt.show()
```

```python
segment_satisfaction = df.groupby(
    "Organization_Segment"
)["Satisfaction"].mean()

print(segment_satisfaction)
```

```python
df["ROI"] = df["Cost_Savings"] / df["AI_Investment"]

segment_roi = df.groupby(
    "Organization_Segment"
)["ROI"].mean()

print(segment_roi)
```
